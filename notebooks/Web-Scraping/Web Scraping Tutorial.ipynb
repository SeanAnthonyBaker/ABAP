```python
# Install necessary packages
!apt update
# Install Google Chrome stable using wget and dpkg
!wget https://dl.google.com/linux/direct/google-chrome-stable_current_amd64.deb
!dpkg -i google-chrome-stable_current_amd64.deb || apt --fix-broken install -y # Install and fix dependencies
# Check if installation was successful by verifying the binary exists
import os
if os.path.exists('/opt/google/chrome/google-chrome'):
    print("Google Chrome binary found after installation.")
else:
    print("Error: Google Chrome binary NOT found after installation.")

!rm google-chrome-stable_current_amd64.deb # Clean up the downloaded file

!pip install selenium webdriver-manager

# Set up Selenium options for headless mode
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Import necessary modules for checking Chrome version
import subprocess
import sys

# --- Access Colab Secrets ---
from google.colab import userdata

try:
    # Retrieve 'EMAIL' and 'PASSWORD' secrets
    user_email = userdata.get('EMAIL')
    user_password = userdata.get('PASSWORD')

    if not user_email or not user_password:
        raise ValueError("Email and/or password secrets not found or accessible. Please set them in Colab Secrets and ensure 'Notebook access' is enabled.")
    print("Successfully retrieved email and password from Colab Secrets.")
except Exception as e:
    print(f"Error accessing Colab Secrets: {e}")
    # Exit the script if credentials cannot be retrieved, as login won't be possible
    sys.exit("Could not retrieve credentials from Colab Secrets. Exiting.")

# --- Selenium Setup ---
options = webdriver.ChromeOptions()
options.add_argument('--headless') # Run browser in headless mode
options.add_argument('--no-sandbox') # Bypass OS security model
options.add_argument('--disable-dev-shm-usage') # Overcome limited resource problems
# Add these additional arguments for robustness
options.add_argument('--disable-gpu')
options.add_argument('--window-size=1920,1080')
options.add_argument('--disable-extensions')
options.add_argument('--disable-features=VizDisplayCompositor') # Additional argument that can help
# Add more arguments that can help with headless execution stability
options.add_argument('--disable-browser-side-navigation')
options.add_argument('--disable-infobars')
options.add_argument('--disable-notifications')
options.add_argument('--disable-logging')
options.add_argument('--log-level=3') # Suppress logs for cleaner output
options.add_argument('--remote-debugging-port=9222') # This helps in certain headless environments
# *** ADD MORE HEADLESS-SPECIFIC ARGUMENTS ***
options.add_argument('--disable-blink-features=AutomationControlled') # Avoid detection as automated browser
options.add_argument('--disable-site-isolation-trials')
options.add_argument('--enable-features=NetworkService,NetworkServiceInProcess')

# --- Specify the executable path explicitly and use webdriver_manager ---
# Specify the path to the google-chrome executable - assuming dpkg installs here
chrome_binary_path = '/opt/google/chrome/google-chrome'
options.binary_location = chrome_binary_path

# Use ChromeDriverManager to get the correct driver.
# ChromeDriverManager will attempt to auto-detect the installed Chrome version
# based on the binary_location if set, or fall back to the latest driver.
webdriver_service = Service(ChromeDriverManager().install())

# --- Initialize the Chrome driver and perform actions ---
driver = None # Initialize driver to None for proper cleanup
try:
    driver = webdriver.Chrome(service=webdriver_service, options=options)
    wait = WebDriverWait(driver, 30) # Increased wait time for dynamic pages and potential network delays

    # --- Navigate to the login page ---
    # We'll use a generic Google login page for demonstration.
    # If you're targeting a specific service, you'll need its precise login URL.
    login_url = 'https://accounts.google.com/ServiceLogin'
    print(f"Navigating to login page: {login_url}")
    driver.get(login_url)

    # --- Enter Email ---
    print("Attempting to find email input field...")
    # Google's email input field often has an ID of 'identifierId'
    try:
        email_input = wait.until(EC.presence_of_element_located((By.ID, 'identifierId')))
        email_input.send_keys(user_email)
        print("Email entered successfully.")
    except Exception as e:
        print(f"Error finding/entering email by ID ('identifierId'): {e}")
        # Fallback for other common locators if 'identifierId' fails
        try:
            email_input = wait.until(EC.presence_of_element_located((By.NAME, 'Email')))
            email_input.send_keys(user_email)
            print("Email entered successfully using By.NAME ('Email').")
        except Exception as e_name:
            print(f"Error finding/entering email by NAME ('Email'): {e_name}")
            raise RuntimeError("Could not find a common email input field. Check page structure.") # Re-raise to stop execution

    # --- Click 'Next' after entering email ---
    print("Attempting to click 'Next' button after email...")
    try:
        # Google's 'Next' button after email often has an ID of 'identifierNext'
        next_button = wait.until(EC.element_to_be_clickable((By.ID, 'identifierNext')))
        next_button.click()
        print("Clicked 'Next' button after email.")
    except Exception as e:
        print(f"Error finding/clicking 'Next' button after email by ID ('identifierNext'): {e}")
        # Fallback for other common locators
        try:
            next_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[@id='identifierNext']//button")))
            next_button.click()
            print("Clicked 'Next' button after email using XPath.")
        except Exception as e_xpath:
            print(f"Error finding/clicking 'Next' button by XPath: {e_xpath}")
            raise RuntimeError("Could not find a common 'Next' button after email. Check page structure.")

    # --- Enter Password ---
    print("Attempting to find password input field...")
    # The password field usually appears on a new screen after clicking next
    try:
        password_input = wait.until(EC.presence_of_element_located((By.NAME, 'password')))
        password_input.send_keys(user_password)
        print("Password entered successfully.")
    except Exception as e:
        print(f"Error finding/entering password by NAME ('password'): {e}")
        # Fallback for other common locators
        try:
            password_input = wait.until(EC.presence_of_element_located((By.ID, 'password')))
            password_input.send_keys(user_password)
            print("Password entered successfully using By.ID ('password').")
        except Exception as e_id:
            print(f"Error finding/entering password by ID ('password'): {e_id}")
            raise RuntimeError("Could not find a common password input field. Check page structure.")

    # --- Click 'Next' (Sign in) after entering password ---
    print("Attempting to click 'Next' (Sign in) button after password...")
    try:
        # The 'Next' or 'Sign in' button after password often has an ID of 'passwordNext'
        signin_button = wait.until(EC.element_to_be_clickable((By.ID, 'passwordNext')))
        signin_button.click()
        print("Clicked 'Next' (Sign in) button after password.")
    except Exception as e:
        print(f"Error finding/clicking 'Next' (Sign in) button after password by ID ('passwordNext'): {e}")
        # Fallback for other common locators
        try:
            signin_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[@id='passwordNext']//button")))
            signin_button.click()
            print("Clicked 'Next' (Sign in) button after password using XPath.")
        except Exception as e_xpath:
            print(f"Error finding/clicking 'Next' (Sign in) button by XPath: {e_xpath}")
            raise RuntimeError("Could not find a common 'Next' (Sign in) button. Check page structure.")

    # --- Verification (Optional) ---
    # After login, you might want to wait for an element that signifies successful login,
    # or check the current URL. For example, if you're redirected to a dashboard:
    print("Waiting for page to load after login...")
    wait.until(EC.url_changes(login_url)) # Wait until the URL changes from the login page
    print(f"Current URL after attempted login: {driver.current_url}")
    print(f"Page title after attempted login: {driver.title}")

    # You can add more actions here, like navigating to the NotebookLM URL
    # or performing specific tasks once logged in.
    # For example, to go to the NotebookLM URL after successful login:
    # driver.get('https://notebooklm.google.com/notebook/142f2ac0-b94c-4a24-ab58-eaf32e2cb583')
    # print(f"Navigated to NotebookLM. Page title: {driver.title}")

except Exception as e:
    print(f"An error occurred during Selenium automation: {e}")
    # You can add more specific error handling or logging here

finally:
    # Ensure the browser is closed even if an error occurs
    if driver:
        driver.quit()
        print("WebDriver closed.")
```
